### NEWTON'S LAW OF UNIVERSAL GRAVITATION

```python
def draw_graph(x, y):
    # Draw the graph
    import matplotlib.pyplot as plt
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitations force and distance')
    plt.show()
```

```python
def generate_F_r():
    # Generate values for r
    r = range(100, 1001, 50)
    # Empty list to store the calulated values of F
    F = []

    # Constant, G
    G = 6.674 * (10**-11)
    # Two masses
    m1 = 0.5
    m2 = 1.5

    #Calculate force and add it to hte list, F
    for distance in r:
        force = G * (m1 * m2) / (distance**2)
        F.append(force)

    # Call the draw_graph funcition
    draw_graph(r, F)
```

```python
generate_F_r()
```
